Q1. Create a line plot of ZN and INDUS in the housing data.

Q2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.
	b. Move the legend to the lower-left corner.
	c. Do the same thing but with horizontal bars.
	d. Move the legend to the upper-right corner.

```python
import matplotlib.pyplot
import numpy as np
df = pd.DataFrame(np.random.rand(5, 3), columns =['a', 'b', 'c'])
ax = df.plot.bar(y=["a","b"])
ax.set_title('Graph of variable a and b', fontsize=21,y=1);
ax.legend(loc='lower left');


ax1 = df.plot.barh(y=["a","b"])
ax1.legend(loc='upper right');
```

Q4. Create a scatter plot of two heatmap entries that appear to have a very positive correlation.
